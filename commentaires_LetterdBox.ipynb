```python
import requests
```

```python
page_to_scrap = requests.get("https://letterboxd.com/film/barbie/")
page_to_scrap.text
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(page_to_scrap.text, "html.parser")
soup
```

```python
comments = soup.findAll("div", attrs={"class": "film-detail-content"})
comments
```

```python
import re
```

```python
regex = re.compile(r"(?<=<p>)(.*?)(?=</p>)")
comments_list = []
for i, comment in enumerate(comments):
    comments_list.append(regex.findall(str(comment)))

comments_list
```

Les commentaires pris ici sont les commentaires les plus likés par d'autres utilisateurs pour voir s'ils sont pertinents. Le site LettedBox les affiche d'ailleurs en haut de la liste des commentaires.
On voit directement en lisant rapidement que les utlisateurs likent plutôt les commentaires drôles donc on ne pourra probablement pas se faire une idée sur le film mais faisons tout de même une analyse de sentiments sur ces commentaires pour commencer.

```python
flattened_comments = [item for sublist in comments_list for item in sublist]
flattened_comments
```

##### Analyse de sentiments

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
analyzer_en = SentimentIntensityAnalyzer()
for comment in flattened_comments:
    vs = analyzer_en.polarity_scores(comment)
    print("{:-<57} {}".format(comment, vs))
```

Avec ce modèle, la plupart des commentaires pour ce film sont classés commme neutre à 100% donc on ne pourra pas se faire une idée subjective du film.
